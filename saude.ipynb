```python
def check_input(f):
    """Checks if the number of inputs is valid"""

    @wraps(f)
    def wrapper(*args, **kwargs):
        if sum([a is not None for a in args]) > 1:
            raise ValueError("At most one of the inputs must be non null")
        return f(*args, **kwargs)

    return wrapper
```
